```python
#points per game vs field goal attempted regression

points_per_game =[29.722892, 26.722892, 29.101266, 27.356322]
field_goals_attempted = [19.060241, 17.518072, 18.797468, 20.241379]

x_values = [points_per_game]
y_values = [field_goals_attempted]

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")


plt.title("Points per Game by Field Goal Attempts")
plt.xlabel("Points Per Game")
plt.ylabel("Field Goals Attempted")

plt.style.use('fivethirtyeight')
plt.show()





# (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
# regress_values = x_values * slope + intercept
# line_eq = "y = " + s(round(slope,2)) + "x + " + str(round(intercept,2))
# plt.scatter(x_values,y_values)
# plt.plot(x_values,regress_values,"r-")
# plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
# plt.show()
```
