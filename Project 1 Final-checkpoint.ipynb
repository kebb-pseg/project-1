```python
# creating a dateframe to show the top 4 players in the MVP race 
#from url https://ak-static.cms.nba.com/wp-content/uploads/sites/46/2021/06/2020-21-NBA-MVP-Voting-Results-Announcement.pdf

# player IDs

# 20 Giannis Antetokounmpo
# 279 Nikola Jokić
# 159 Joel Embiid
# 124 Stephen Curry


# MVP Votes of 4 players
mvp_votes_data = {20: 348, 279: 971, 159: 586, 124: 453}
mvp_votes_df = pd.DataFrame(list(mvp_votes_data.items()), columns=["player_id", "mvpvotes"])

# Display the DataFrame
mvp_votes_df
```

```python
import pandas as pd
import requests
import json
import pprint

url = "https://api-nba-v1.p.rapidapi.com/players/statistics"
headers = {
    "X-RapidAPI-Key": api_key,
    "X-RapidAPI-Host": "api-nba-v1.p.rapidapi.com"
}
#pramaters for API
player_ids = [20, 279, 159, 124]
season = 2021
game_statistics = []

#looping player id to get games statistics for the season
#skip the game data if the player does not play in
for player_id in player_ids:
    querystring = {"id":player_id,"season":season}
    try:
        response = requests.get(url, headers=headers, params=querystring)
        statistics = response.json()["response"]
        #print(json.dumps(statistics, sort_keys=True, indent=4))
        for s in statistics:
            #if the play not playing, the position data is null
            if (s["pos"] is not None):
                game_statistics.append({                
                    "player_id": s["player"]["id"], 
                    "name" : s["player"]["lastname"],                
                    "points": s["points"],
                    "fgm":s["fgm"],
                    "fga":s["fga"],
                    "fgp":float(s["fgp"]),
                    "ftm":s["ftm"],
                    "fta":s["fta"],
                    "ftp":float(s["ftp"]),
                    "tpm":s["tpm"],
                    "tpa":s["tpa"],
                    "tpp":float(s["tpp"]),
                    "offReb":s["offReb"],
                    "defReb":s["defReb"],
                    "assists":s["assists"],
                    "pFouls": s["pFouls"],
                    "steals": s["steals"],
                    "turnovers": s["turnovers"],
                    "blocks": s["blocks"],
                    "plusMinus": float(s["plusMinus"]),
                    "min": float(s["min"].split(":")[0]) + float(s["min"].split(":")[1]) /60                
                })
    except:        
        print("Error on getting data from API call")        
game_df = pd.DataFrame(game_statistics)
# the mean values group by player id
mean_df = game_df.groupby(['player_id','name'], as_index=False, sort=False).mean()
# the game count group by player id
count_df = game_df.groupby(['player_id','name'], sort=False)["min"].count().to_frame("games")["games"]
# join the two dataFrame: mean_df and count_df 
statistics_df = pd.merge(count_df,mean_df,on=['player_id','name'],how='left')
# join with mvp_votes dataFrame
statistics_df = pd.merge(statistics_df,mvp_votes_df,on=['player_id'],how='left')
# final statistics data by the player
statistics_df
```
